# The ghi for day 18-05 are concatenated into one dataframe

```python
df_ghui=pd.concat([ghi18,ghi19,ghi21,ghi22,ghi23,ghi24,ghi25,ghi26,ghi27,ghi28,ghi29,ghi30,ghi31,ghi01,ghi05])
df_ghui
```

```python
df_ghui.rename(columns = {'Time':'T-Match', 'GHI':'GHI'}, inplace = True)
df_ghui
```

# The final updated dataset for Regresion model

```python
# Now concatenating 18,19,21,22,23,24,25,26,27,27,29,30,31,01, & 05 days data for both lux, ghui we will merge them
ghi_list = df_ghui['GHI'].tolist()
match_list = df_ghui['T-Match'].tolist()
df_lux['T-Match'] = match_list
df_lux['GHI'] = ghi_list

#check match
check = df_lux['Time'].equals(df_lux['T-Match'])
print(f"Time matching {check}")

df_lux
```

```python
df_lux['Time'].equals(df_lux['T-Match'])
```

```python
df_final = df_lux.drop(['T-Match','Lux0 Regression'], axis=1)
df_final
```

```python
df_final.rename(columns = {'Lux0 Regression':'Lux0', 'Lux1 Regression':'Lux1', 'Lux2 Regression':'Lux2', 'Lux3 Regression':'Lux3', 'Lux4 Regression':'Lux4', 'Lux5 Regression':'Lux5', 'Lux6 Regression':'Lux6', 'Lux7 Regression':'Lux7', 'Lux8 Regression':'Lux8',}, inplace = True)
df_final.drop_duplicates(subset=['Time'],ignore_index=True, inplace=True)
df_final
```
